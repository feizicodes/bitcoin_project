```python
#save the data in a SQLite database

import requests
import sqlite3

# API-Anfrage
url = "https://www.alphavantage.co/query?function=DIGITAL_CURRENCY_DAILY&symbol=BTC&market=EUR&apikey=demo"
response = requests.get(url)
data = response.json()

# Tägliche Daten extrahieren
timeseries = data["Time Series (Digital Currency Daily)"]

# Verbindung zu SQLite
conn = sqlite3.connect("crypto_data.db")
cur = conn.cursor()

# Tabelle anlegen
cur.execute('''
    CREATE TABLE IF NOT EXISTS crypto_prices (
        date TEXT PRIMARY KEY,
        open REAL,
        high REAL,
        low REAL,
        close REAL,
        volume REAL
    )
''')

# Daten einfügen
for date, values in timeseries.items():
    cur.execute('''
        INSERT OR REPLACE INTO crypto_prices (date, open, high, low, close, volume)
        VALUES (?, ?, ?, ?, ?, ?)
    ''', (
        date,
        float(values["1. open"]),
        float(values["2. high"]),
        float(values["3. low"]),
        float(values["4. close"]),
        float(values["5. volume"])
    ))

# Speichern & schließen
conn.commit()
conn.close()
```
